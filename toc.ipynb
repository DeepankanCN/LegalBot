```python
def chunk_with_overlap(pages, overlap=500):
    """
    Create merged chunks with overlap from a list of page strings.

    :param pages: List of strings, where each string represents a page of a PDF document.
    :param overlap: Number of overlapping characters between chunks.
    :return: List of strings, each containing the merged current_chunk and next_chunk.
    """
    merged_chunks = []

    for i in range(len(pages) - 1):
        current_page = pages[i]
        next_page = pages[i + 1]

        # Determine the start and end of the current chunk
        chunk_end = len(current_page) - overlap
        if chunk_end < 0:
            chunk_end = len(current_page)

        current_chunk = current_page#[:chunk_end]
        next_chunk = next_page[:overlap]

        # Merge the chunks
        merged_chunk = current_chunk + "\n" +next_chunk

        # Append the merged chunk to the list
        merged_chunks.append(merged_chunk)

    # Optionally handle the last page separately
    if len(pages) > 1:
        merged_chunks.append(pages[-1])

    return merged_chunks



overlap = 500
merged_chunks = chunk_with_overlap(extracted_text, overlap)
k=2
# Print results
for i, chunk in enumerate(merged_chunks):
    print(f"Chunk {i + 1} (length {len(chunk)}):")
    print(chunk)
    print()
    break
    # if k>2:
    #     break
    # k=k+1
```
